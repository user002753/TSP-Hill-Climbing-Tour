```python
import math
import random
import matplotlib.pyplot as plt
```

```python
#calculating euclidean distance
def distance(city1,city2):
  return math.sqrt((city1[0]-city2[0])**2+(city1[1]-city2[1])**2)
```

```python
#calculating total distance of a tour
def tour_distance(tour,cities):
  dist =0
  for i in range(len(tour)):
      dist += distance(cities[tour[i]],cities[tour[(i+1)%len(tour)]])
  return dist
```

```python
#to swap two cities in a tour
def swap_cities(tour):
  new_tour=tour[:]
  i,j=random.sample(range(len(tour)),2)
  new_tour[i],new_tour[j]=new_tour[j],new_tour[i]
  return new_tour
```

```python
#hill climbing algorithm
def hill_climbing(cities,max_iter):
  tour=list(range(len(cities)))
  random.shuffle(tour)
  current_distance=tour_distance(tour,cities)
  for i in range(max_iter):
    new_tour=swap_cities(tour)
    new_distance=tour_distance(new_tour,cities)
    if new_distance<current_distance:
      tour=new_tour
      print(f"Iteration {i+1}:Distance={current_distance:.2f}")
  return tour,current_distance
```

```python
#to plot the tour
def plot_tour(cities,tour):
  tour_cities=[cities[i] for i in tour]+[cities[tour[0]]]
  xs,ys=zip(*tour_cities)
  plt.figure(figsize=(8,6))
  plt.plot(xs,ys,'o-',color='blue')
  for i,city in enumerate(tour):
    plt.text(cities[city][0]+0.5,cities[city][1]+0.5,str(city),fontsize=12)
  plt.title("TSP Hill Climbing Tour")
  plt.xlabel("X coordinate")
  plt.ylabel("Y coordinate")
  plt.show()


```

```python
#Main
if __name__ == "__main__":
  cities=[(random.randint(0,100),random.randint(0,100)) for _ in range(10)]
  print("cities:",cities)
  best_tour,best_distance=hill_climbing(cities,50)
  print("Best Tour:",best_tour)
  print("Best Distance:",best_distance)
  plot_tour(cities,best_tour)
```
